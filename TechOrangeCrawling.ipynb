```python
import requests
from bs4 import BeautifulSoup as bs
import lxml
from IPython.display import display, HTML
import pdfkit
import bs4
```

```python
#page = requests.get('https://buzzorange.com/techorange/2017/06/20/start-from-zero-learning-big-data/')
#page = requests.get('https://buzzorange.com/techorange/2017/08/30/how-to-work-on-ai/')
page = requests.get('https://buzzorange.com/techorange/2017/08/21/the-best-ai-lesson/')

article = bs(page.text,'lxml')
#display(HTML(str(article)))
```

```python
mainArt = article.find('article')
```

```python
#mainArt.find(class_='signature').decompose()
mainArt.find(class_='fb-count').decompose()
mainArt.find(class_='byline').decompose()
for item in mainArt.find_all('img',class_='aligncenter'):
    item.decompose()
```

```python
startDecompose = False
for item in mainArt.find(class_='entry-content').children:
    if item==mainArt.find(id='share_group_btns'):
        startDecompose = True
    if startDecompose:
        if type(item)!=bs4.NavigableString:
            item.decompose()
```

```python
title = str(mainArt.find('h1').contents[0])
```

```python
newPage = bs('<html><head><title></title><meta charset="UTF-8"></head><body></body></html>','lxml')
newPage.title.insert(0,title)
newPage.body.insert(0,mainArt)
```

```python
with open(title+'.html','w',encoding='utf-8') as file:
    file.write(newPage.prettify())
    
pdfkit.from_file(title+'.html', title+'.pdf')
```
